```python
#Creates a piechart from the below data
#Call libraries
import pandas as pd
import seaborn 
import matplotlib.pyplot as plt

# declaring data
data = [44, 45, 40, 41, 39]
keys = ['Class 1', 'Class 2', 'CLass 3', 'Class 4', 'Class 5']
  
# declaring exploding pie
explode = [0, 0.1, 0, 0, 0]
# define Seaborn color palette to use
palette_color = seaborn.color_palette('dark')
  
# plotting data on chart
plt.pie(data, labels=keys, colors=palette_color,
        explode=explode, autopct='%.0f%%')
  
# displaying chart
plt.show()


```

**Normalized losses based on No of wheels and body style**
<p>Findings: As we understand the normalized loss which is the average loss payment per insured vehicle is calculated with many features of the cars which includes body style and no. of doors. Normalized losses are distributed across different body style but the two door cars has more number of losses than the four door cars.</p>

```python
pd.pivottable(automobile.index=['body-style','num-of-doors'],values='normalized_losses').plot(kind='bar',color='blue')
plt.title('Number of losses based on Body style and number of doors')
plt.ylabel('Normalized losses')
plt.xlabel('Body style and No. of doors')
```
