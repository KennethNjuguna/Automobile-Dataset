```python
# Make simple linearplot with the dataset

import seaborn as sns
sns.set_style('whitegrid')
sns.lmplot(x='engine-size', y='stroke', data=automobile)
```

```python
#This Plot Introduces the Hue parameter chooses which columns paremeter should be used to colour on the plot.
#Below by passing the hue parameter on drive-wheels the plot colours differently.

import seaborn as sns
import matplotlib as plt
sns.set_style('whitegrid')
sns.lmplot(y="stroke", x="engine-size", hue="drive-wheels", markers=['D','H','X'], data=automobile)
```

```python
#implement a distribution plot on The automobile dataset
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')
automobile=pd.read_csv('/kaggle/input/automobile/automobile.csv')
sns.displot(automobile['engine-size'], kde= False, color='red')
sns.displot(automobile['horsepower'], kde=False, color='blue', bins=50)
```

```python
#Creates a piechart from the below data
#Call libraries
import pandas as pd
import seaborn 
import matplotlib.pyplot as plt

# declaring data
data = [44, 45, 40, 41, 39]
keys = ['Class 1', 'Class 2', 'CLass 3', 'Class 4', 'Class 5']
  
# declaring exploding pie
explode = [0, 0.1, 0, 0, 0]
# define Seaborn color palette to use
palette_color = seaborn.color_palette('dark')
  
# plotting data on chart
plt.pie(data, labels=keys, colors=palette_color,
        explode=explode, autopct='%.0f%%')
  
# displaying chart
plt.show()


```
